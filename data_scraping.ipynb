```python
from datetime import datetime
start_time = datetime.now()
print start_time 
import requests
from bs4 import BeautifulSoup
import numpy as np
import csv
import pandas as pd
import re
```

```python
#access the recipe index on the foodnetwork website
#get the url for the different pages containing recipes beginning with each alphabet and number
def get_links(response):
    soup = BeautifulSoup(response.text, "html5lib")
    #print(soup)
    links = soup.findAll('a')
    #print(links)
    ud_df=[]
    for link in links:
        #print(link)
        try:
            if link.get('href').startswith('//www.foodnetwork.com/recipes/a-z/'):
                print(link.get('href'))
                ud_df.append(link.get('href'))
                print("done updating")
        except:
            continue
        
    return ud_df

page = requests.get("http://www.foodnetwork.com/content/food-com/en/recipes/a-z.html",allow_redirects=False)
res=get_links(page)
```

```python
#Get the total number of pages worth of recipes on the website
def num_pages(res1):
    list_link=[]
    print res1 
    for link1 in res1:
        link_max=0
        pp=requests.get('http:'+link1)
        soup=BeautifulSoup(pp.text,"html5lib")
        links_pp=soup.findAll('li', class_="o-Pagination__a-ListItem")
        max_arr=[]
        for link_pp in links_pp:
            links_all=link_pp.find_all("a", class_="o-Pagination__a-Button ")
            array1=[]
            for link_all in links_all:
                if link_all:
                    array1.append(int(str(link_all.get('href'))[-2:].replace('/','')))
            #print array1
            #print "that was array"
            if array1:
                max_arr.append(max(array1))
                #print "that was max"
                
            links_all1=link_pp.find_all("a", class_="o-Pagination__a-Button is-Active ")
            array2=[]
            for link_all1 in links_all1:
                if link_all1:
                    array2.append(int(str(link_all1.get('href'))[-2:].replace('/','')))
            #print array2
            #print "that was array"
            if array2:
                max_arr.append(max(array2))
                #print max(array2) 
        if max_arr:
            link_max=max(max_arr)
                #print "that was max"
        list_link.append([link1,link_max])
    return list_link
print(res)
page_info=num_pages(res)
```

```python
#get the different recipes on each page of the foodnetwork website
def get_recipes(pinfo):
    link_pages=[]
    rl=[]
    cnt=0
    for row in pinfo:
        for i in range(row[1]):
            link_pages.append(row[0]+"/p/"+str(i+1))
    for link in link_pages:
        print link
        page=requests.get('http:'+link)
        soup=BeautifulSoup(page.text,"html5lib")
        links_page=soup.findAll('li', class_="m-PromoList__a-ListItem")
        for link in links_page:
            #print link
            #print "*************************************************************"
            recipe_links=link.findAll('a')
            for recipe_link in recipe_links:
                rl.append(recipe_link.get('href'))
        #print len(rl)
    cnt=len(rl)
    print cnt
    return rl
            
recipelink=get_recipes(page_info) 
```

```python
#get data for each recipe
#title, chef, number of reviews, rating, ingredients, cooktime, recipe level, recipe category etc.
import urllib2
import json

def get_data(rcps):
    print "entered"
    j=0
    i=0
    recipe_data=[]
    rcp_not_found=[]
    for rcp in rcps:
        cat=""
        rev=""
        rate=""
        rcp='http:'+rcp
        print(rcp)
        req = urllib2.Request(rcp)
        req.add_header('Accept', 'application/json')
        
        try:
            #print "entered"
            res = urllib2.urlopen(req)
        except:
            j=j+1
            rcp_not_found.append(rcp)
            continue
            
        soup=BeautifulSoup(res,'html5lib')
        #print soup
        
        cat1=soup.findAll('div',class_="o-Capsule__m-TagList m-TagList")
        #.findAll('div',class_="o-Capsule__m-TagList m-Tag")
        
        for row1 in cat1:
            cat2=row1.findAll('a',class_="o-Capsule__a-Tag a-Tag")
            for row2 in cat2:
                cat+=","+row2.text   
        
        ings=soup.findAll('div',class_="o-Ingredients__m-Body")
        #print ings
        for ing in ings:
            ing_list=""
            ing_l=ing.findAll("li",class_="o-Ingredients__a-ListItem")
            for ingr in ing_l:
                ingr1=ingr.findAll('label')
                for each in ingr1:
                    #print each.text
                    ing_list+="\n"+each.text
               
        levels=soup.findAll('dd',class_="o-RecipeInfo__a-Description")
        lvl=""
        for level in levels:
            if 'easy' in level.text.lower():
                lvl=level.text.lstrip().rstrip()
            elif 'intermediate' in level.text.lower():
                lvl=level.text.lstrip().rstrip()
            elif 'advanced' in level.text.lower():
                lvl=level.text.lstrip().rstrip()
        titles=soup.findAll('title')
        for title in titles:
            temp=title.text.split('|')
            rec_title=temp[0]
            chef=temp[1]
        try:
            other_info=soup.findAll('script', type="application/ld+json")
        except:
            continue
        data_list=list(other_info[0])
        data_dict=dict((k,2) for k in data_list)
        for key,value in data_dict.items():
            key_list=key.split(",")
            key_dict={}
            for row1 in key_list:
                rat1=[]
                rat1=row1.split(": ")
                if "ratingValue" in rat1[0]:
                    #print rat1
                    #print rat1[1]
                    rate=rat1[1]
                    break
            for row2 in key_list:
                rat2=[]
                rat2=row2.split(": ")
                
                if "reviewCount" in rat2[0]:
                    #print rat2
                    if " " or "{" or "\n" in rat2[1]:
                        rat2[1]=rat2[1].replace(" ","")
                        rat2[1]=rat2[1].replace("}","")
                        rat2[1]=rat2[1].replace("\n","")
                    #print rat2[1]
                    rev=rat2[1]
                    
            for row3 in key_list:
                rat3=row3.split(": ")
                if "totalTime" in rat3[0]:
                    time_list=re.findall(r'[A-Za-z]+|\d+', rat3[1])
                    cook_time=time_list[7]+time_list[8]+time_list[9]+time_list[10]
                    break
        #print type(rec_title),type(chef),type(rate),type(rev),type(cook_time),type(lvl),type(ing_list),type(cat),type(rcp)
        #print rec
        #print "*************************************************************************"
        try:
            rec=rec_title+" sep "+chef+" sep "+rate+" sep "+rev+" sep "+cook_time+" sep "+lvl+" sep "+ing_list+" sep "+cat+" sep "+rcp   
            recipe_data.append(rec)
        except:
            continue
        i=i+1
        if i%1000 == 0:
            print "%d recipes done" %i
            print "*****************************************"
    print j
    return recipe_data
        
       
```

```python
#store the data in a list
rcp_data=get_data(recipelink[60000:])
```
